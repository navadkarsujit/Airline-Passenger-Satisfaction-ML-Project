```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('dark_background')
```

### Import Data & merge

```python
d1 = pd.read_csv('Airline-Passenger-train.csv')
d2 = pd.read_csv('Airline-Passenger-test.csv')

ds = d1._append(d2, ignore_index=True)

ds = ds.iloc[:, 2:]

ds
```

```python
ds.info()
```

```python
ds.isna().sum()
```

## Handle Missing Data

```python
ds["Arrival Delay in Minutes"].fillna(0, inplace=True)
```

```python
ds.describe().T
```

## Data Viz

```python
fig, ax = plt.subplots(3, 2, figsize=(15, 10))

scols = ds.select_dtypes(object).columns

for ax, col in zip(ax.flatten(), scols):
    sns.countplot(ds, x=col, ax=ax, hue='satisfaction')
```

```python
ncols = ds.select_dtypes('number').columns

fig, ax = plt.subplots(6, 3, figsize=(15, 35))

for ax, col in zip(ax.flatten(), ncols):
    sns.violinplot(ds, x=col, ax=ax)
```

## Data Transformation

```python
from scipy.stats import boxcox

ds["Flight Distance"] = boxcox(ds["Flight Distance"])[0]

ds["Arrival Delay in Minutes"] = np.sqrt(np.log1p(ds["Arrival Delay in Minutes"]))

ds["Departure Delay in Minutes"] = np.sqrt(np.log1p(ds["Departure Delay in Minutes"]))
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

for col in scols:
    ds[col] = LabelEncoder().fit_transform(ds[col])
    
Y = ds["satisfaction"]
    
ds = pd.DataFrame(StandardScaler().fit_transform(ds.iloc[:, :-1]), columns=ds.columns[:-1])

ds["Y"] = Y

ds
```

```python
plt.subplots(1, 1, figsize=(20, 15))

sns.heatmap(ds.corr(), annot=True)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(ds.iloc[:, :-1], ds.iloc[:, -1], train_size=.9)

xTrain.shape, yTest.shape
```

## Model

```python
from tensorflow.keras.layers import Dense, Dropout, LeakyReLU, BatchNormalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam, SGD, RMSprop
from tensorflow.keras.losses import CategoricalCrossentropy

model = Sequential([
    Dense(xTrain.shape[1]),
    LeakyReLU(),
    BatchNormalization(),
    Dense(16),
    LeakyReLU(),
    BatchNormalization(),
    Dropout(.2),
    Dense(32),
    LeakyReLU(),
    BatchNormalization(),
    Dropout(.5),
    Dense(16),
    LeakyReLU(),
    BatchNormalization(),
    Dense(32),
    LeakyReLU(),
    BatchNormalization(),
    Dropout(.2),
    Dense(16),    
    LeakyReLU(),
    BatchNormalization(),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=.002), metrics=['Accuracy'])

history = model.fit(xTrain, yTrain, batch_size=256, validation_data=(xTest, yTest), epochs=150)
```

## Model Evaluation

```python
loss = history.history['loss']
acc = history.history['Accuracy']

fig, ax = plt.subplots(1, 2, figsize=(20, 5))

ax[0].plot(acc)
ax[0].set_xlabel("Accuracy")

ax[1].plot(loss)
ax[1].set_xlabel("Loss")
```

```python
model.evaluate(xTest, yTest)
```

```python
yPred = list(map(lambda x: .0 if x < 0.5 else 1., np.ravel(model.predict(xTest))))
```

```python
pd.Series(yPred).value_counts()
```

```python
from sklearn.metrics import classification_report, accuracy_score

print (classification_report(yTest, yPred, zero_division=True))
```
